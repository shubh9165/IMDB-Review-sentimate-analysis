```python
import pandas as pd
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
documents=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
print(documents)
```

```python
vocab_size=10000
```

```python
one_hot_rep=[one_hot(words,vocab_size) for words in documents]
```

```python
print(one_hot_rep)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
```

```python
set_length=10
embedded_docs=pad_sequences(one_hot_rep,padding='post',maxlen=set_length)
print(embedded_docs)
```

```python
model=Sequential()
model.add(Embedding(vocab_size,10,input_length=set_length))
model.compile('adam','mse')
```

```python
vec=model.predict(embedded_docs)
```
